```python
import pandas as pd
import sklearn
```

```python
import numpy as np
```

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
print("Classes to predict:",data.target_names)
```

```python
#X is the collection of sepal and petal length and width
X= data.data
#y is the names of the classes to which Iris belongs to
y= data.target
```

```python
print(X[0])
```

```python
print(y[145])
```

```python
X[:4]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=47)
```

```python
print("{} => {}".format(X_train[0],y_train[0]))
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy score on training data:",accuracy_score(y_true=y_train,y_pred=clf.predict(X_train)))
print("Accuracy score on test data:",accuracy_score(y_true=y_test,y_pred=y_pred))
```

```python
clf = DecisionTreeClassifier(criterion='entropy',min_samples_split=10)
```

```python
clf.fit(X_train,y_train)
print("Accuracy score on training data:",accuracy_score(y_true=y_train,y_pred=clf.predict(X_train)))
print("Accuracy score on test data:",accuracy_score(y_true=y_test,y_pred=y_pred))
```
